```python
import cv2
```

```python
img1 = cv2.imread("../DATA/dog_backpack.png")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.imread("../DATA/watermark_no_copy.png")
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
```

```python
def show_img(img):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
show_img(img2)
```

```python
print(img1.shape)
print(img2.shape)
```

```python
# blending images of same size
img1_1 = cv2.resize(img1,(1200,1200))
img2_1 = cv2.resize(img2,(1200,1200))
```

```python
show_img(img1_1)
```

```python
blended_img = cv2.addWeighted(img1_1,0.8,img2_1,0.1,0)
show_img(blended_img)
```

```python
# overlay  different sizes
# small image on large image , reassign the numpy array value
```

```python
img2_1 = cv2.resize(img2,(600,600))
```

```python
img1_1[0:img2_1.shape[0],0:img2_1.shape[1]] = img2_1
```

```python
show_img(img1_1)
```

```python
img2gray = cv2.cvtColor(img2_1, cv2.COLOR_RGB2GRAY)
```

```python
show_img(img2gray)
```

```python
plt.imshow(img2gray, cmap="gray")
```

```python
# mask takes only the white portion and removes all black
```

```python
mask_inv = cv2.bitwise_not(img2gray)
```

```python
plt.imshow(mask_inv, cmap="gray")
```

```python
mask_inv.shape
```

```python
img2w = np.full(img2_1.shape,255)
```

```python
bk = cv2.bitwise_or(img2w, img2w, mask=mask_inv)
```
