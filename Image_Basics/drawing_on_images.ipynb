# Drawing on image with the mouse
- we can use callback to connect event functions to opencv


```python
img = np.zeros(shape=(500,500,3))
```

```python
def draw_circle(event,x,y,flags,params):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),100,color=(255,0,0), thickness=-1)
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y),100,color=(0,255,255), thickness=-1)
```

```python
#connecting callback functions

cv2.namedWindow(winname="drawing")
cv2.setMouseCallback("drawing",draw_circle)
while True:
    cv2.imshow("drawing",img)
    if cv2.waitKey(10) & 0xFF == ord("r"): 
        break
    
cv2.destroyAllWindows()
```

## Build Rectangle by dragging

```python
img2 = np.zeros(shape=(500,500,3))
```

```python
window = "my_rec"
```

```python
drawing = False
ix,iy = 0,0
def draw_rectangle(event,x,y,flags,params):
    global drawing, ix,iy, img2
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix,iy = x,y
        cv2.circle(img,(x,y),1,color=(255,0,0), thickness=-1)
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            cv2.rectangle(img2,pt1=(ix,iy),pt2=(x,y),color=(255,0,0),thickness=-1)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(img2,pt1=(ix,iy),pt2=(x,y),color=(255,0,0),thickness=-1)
```

```python
cv2.namedWindow(winname="my_rec")
cv2.setMouseCallback("my_rec",draw_rectangle)
while True:
    cv2.imshow("my_rec",img2)
    if cv2.waitKey(1) & 0xFF == ord("r"): 
        break
    
cv2.destroyAllWindows()
```
